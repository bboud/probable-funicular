```python
import pyart
import matplotlib.pyplot as plt
import numpy as np

radar = pyart.io.read('KTLX_N0U/KOUN_SDUS54_N0UTLX_201305312337')

#mask out last 10 gates of each ray, this removes the "ring" around the radar.
radar.fields["velocity"]["data"][:, -10:] = np.ma.masked

# exclude masked gates from the gridding
gatefilter = pyart.filters.GateFilter(radar)
gatefilter.exclude_transition()
gatefilter.exclude_masked("velocity")

# perform Cartesian mapping, limit to the reflectivity field.
grid = pyart.map.grid_from_radars(
    (radar,),
    gatefilters=(gatefilter,),
    grid_shape=(1, 241, 241),
    grid_limits=((100, 100), (-230000.0, 230000.0), (-230000.0, 230000.0)),
    fields=["velocity"],
)

# create the plot
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111)
ax.imshow(grid.fields["velocity"]["data"][0].data, origin="lower")
plt.show()
```

```python
import pyfeats
masked = grid.fields["velocity"]["data"][0]
f = masked.data
f = f / np.linalg.norm(f)

plt.imshow(f, origin="lower")
```
